```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Label prediction (SVM and Node2Vec)

```python
labels = ['0', '1', '2', '3', '4']
y_test = [4.960e+02, 1.820e+02, 2.540e+02, 5.104e+03, 4.000e+00]
y_pred = [2851., 1378.,  741., 4526.,  151.]
# put in a dataframe
df = pd.DataFrame({'labels': labels, 'y_test': y_test, 'y_pred': y_pred})
df.plot(x="labels", y=["y_test", "y_pred"], kind="bar", rot=0)
#bigger plot
plt.xlabel('Labels')
plt.ylabel('Count')
plt.title('Comparison between y_test (groundtruth) and y_pred (predicted)')
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1), title="Labels")
#log log
plt.yscale('log')
#save
plt.savefig('assets/y_test_vs_y_pred (SVM).png', bbox_inches='tight')
```
